Numpy Attributes

```python
import numpy as np
```

```python
a1=np.arange(10)
print(a1)
a2=np.arange(12,dtype=float).reshape(3,4)
print(a2)
a3=np.arange(8).reshape(2,2,2)
print(a3)
```

```python
print(a1.ndim)
print(a2.ndim)
print(a3.ndim)
print(a1.shape)
print(a2.shape)
print(a3.shape)
```

```python
a1.itemsize
```

```python
a2.itemsize
```

```python
b1=np.arange(10,dtype=np.int32)
print(b1)
b1.itemsize
```
