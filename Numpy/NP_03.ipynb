Array-Function

```python
arr=np.random.random((3,3))
arr=np.round(arr*100)
arr
```

```python
# max/min/sum/prod
print("Max:",np.max(arr))
print("Min:",np.min(arr))
print("Sum:",np.sum(arr))
print("Prod:",np.prod(arr))
```

```python
# we also find row-wise and column-wise min max sum etc.
print("Row-wise sum:",np.sum(arr,axis=1))
print("Column-wise sum:",np.sum(arr,axis=0))
print("Row-wise max:",np.max(arr,axis=1))
print("Column-wise max:",np.max(arr,axis=0))
```

```python
print(np.mean(arr))
print(np.std(arr))
print(np.median(arr))
```

```python
# dot product
a4=np.array([[1,2,3],[4,5,6]])
a5=np.array([[7,8],[9,10],[11,12]])
np.dot(a4,a5)
```
