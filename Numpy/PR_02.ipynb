- Given two random $3 \times 3$ arrays A and B:
- Compute the element-wise product.
- Compute the matrix dot product using np.dot or the @ operator.
- Add a comment explaining the mathematical difference between the two operations.

```python
import numpy as np

A=np.random.random((3,3))
print(A.shape)
B=np.random.random((3,3))
print(B.shape)
```

```python
a1=np.array([1,2,3,4])
b1=np.array([[2],[2],[2],[2]])
```

```python
a1@b1
```

```python
np.dot(a1,b1)
```

```python
arr1=A@B
arr1
```

```python
A = np.array([[1,2],[3,4]])
B = np.array([[5,6],[7,8]])
```

```python
print(A @ B)
print(np.dot(A,B))
```

```python
A = np.ones((2,3,4))
B = np.ones((2,4,5))
print(A @ B)
print(np.dot(A,B))
```

```python
# @ is strict matrix multiplication, np.dot() is dimension-dependent and can behave differently for higher-dimensional arrays.
```
